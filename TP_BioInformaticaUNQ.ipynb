```python
# Para prototipar el proyecto
from Bio.PDB import MMCIFParser,NeighborSearch

# Demo file
demo_file = "3U8K.cif"

# Distancia de busqueda
cutoff_distance = 4

def load_structure():
  parser = MMCIFParser(QUIET=True)
  structure = parser.get_structure("prot",demo_file)
  return structure

def search(ligando_atoms,structure):
  residues = []
  atoms_list = list(structure.get_atoms()) #TODO: esta bien convertir a lista?
  neighbor_search = NeighborSearch(atoms_list) # lista de atomos ?

  for atom in ligando_atoms: # --> en el video lo hace para un solo atomo de hierro
    #print(f"atom : {atom}")
    #print(f"atom.get_coord() : {atom.get_coord()}")
    neighbors = neighbor_search.search(atom.get_coord(), cutoff_distance)  # --> confirmar con pymol | chimera
    residues.extend(neighbors)
    #TODO: lista de residuos o residuo de atomos


  # filtrar neighbors para que solo sean residuos de la proteina
  #print(f"residues : {residues}")
  return residues

#coordenadas de atomos de los ligandos de la proteina
def ligando_coords(structure):
  ligando_atoms = []

  for model in structure:
    for chain in model:
      for residue in chain:  # HETATM  -> son los ligandos
        hetflag = residue.id[0]
        resname = residue.resname

        #agregar solo ligandos
        if hetflag != " " and resname != "HOH":     # -> check si no usar HOH
          for atom in residue:
            ligando_atoms.append(atom)

  #print(f"ligando_atoms:{ligando_atoms}")
  return ligando_atoms


# ----------- "MAIN" ----------------------

def residuos_de_la_proteina():
  # Cargo la estructura
  structure = load_structure()

  # Busco las coordenadas de ligandos
  ligando_atoms = ligando_coords(structure)

  # Busco los residuos
  residues = search(ligando_atoms,structure)

  print(f"residues : {residues}")
  return residues


residuos_de_la_proteina()
```
